```python
import custom_box_plot as cbp
import numpy as np
```

```python
# Prepare data
fig_ylabels = ['SSP370', 'SSP585']

scenario_1_labels = ['Model_01', 'Model_02', 'Model_03', 'Model_04', 'Model_05']
model_1_data = np.random.rand(2)
model_2_data = np.random.rand(5)
model_3_data = np.random.rand(1)
model_4_data = np.random.rand(3)
model_5_data = np.random.rand(4)
scenario_1_data = [model_1_data, model_2_data, model_3_data, model_4_data, model_5_data]

scenario_2_labels = ['Model_A', 'Model_B', 'Model_C', 'Model_D', 'Model_E']
model_a_data = np.random.rand(5)
model_b_data = np.random.rand(2)
model_c_data = np.random.rand(4)
model_d_data = np.random.rand(1)
model_e_data = np.random.rand(3)
scenario_2_data = [model_a_data, model_b_data, model_c_data, model_d_data, model_e_data]
```

```python
cbp.custom_box_plot(
    data=[scenario_1_data, scenario_2_data],
    labels=[scenario_1_labels, scenario_2_labels],
    fig_save_as='./custom_box_plot_example.png',
    fig_ylabels=fig_ylabels,
    fig_title='Custom Box Plot Example',
)
```
